```r
library(ggplot2)
library(vegan)
library(dplyr)
library(scales)
library(grid)
library(reshape2)
library(phyloseq)
#install.packages("ggfortify")
library('ggfortify')
packageVersion('ggfortify')
library(data.table)
```

```r
otu_table<-read.csv("OTU_table_Kelleyetal.tsv",sep = '\t',check.names=FALSE)
denovo<-read.csv("completeRF.csv",sep = ',',check.names=FALSE)
metafile <- read.csv("pcos_mapping_Treat.tsv",sep = '\t',check.names=FALSE)
```

```r
otu_table
denovo
metafile
```

```r
all <- merge(otu_table,denovo,by="OTU_ID")
all_re <- all[,c(121,1:120)]
all_re <- select(all_re,-c(OTU_ID))
all_rename <- all_re %>% 
  rename(
    OTU_ID = new.col
  )
```

```r
all_rename
```

```r
#all_t <- as.data.frame(t(as.matrix(all_rename)))
#all_t
all_t = setNames(data.frame(t(all_rename[,-1])), all_rename[,1])
all_t
```

```r
all_t <- setDT(all_t,keep.rownames = "OTU_ID")[]
all_t
```

```r
merge_otu_treat <- merge(all_t,metafile,by="OTU_ID")
merge_otu_treat
```

```r
time2 <- subset(merge_otu_treat, Time == '2')
time5 <- subset(merge_otu_treat, Time == '5')
time1 <- subset(merge_otu_treat, Time == '1')
time3 <- subset(merge_otu_treat, Time == '3')
time4 <- subset(merge_otu_treat, Time == '4')
```

```r
time5
```

```r
write.table(time2, file='RF_time2.tsv', quote=FALSE, sep='\t', col.names = TRUE, row.names = FALSE)
write.table(time5, file='RF_time5.tsv', quote=FALSE, sep='\t', col.names = TRUE, row.names = FALSE)
write.table(time1, file='RF_time1.tsv', quote=FALSE, sep='\t', col.names = TRUE, row.names = FALSE)
write.table(time3, file='RF_time3.tsv', quote=FALSE, sep='\t', col.names = TRUE, row.names = FALSE)
write.table(time4, file='RF_time4.tsv', quote=FALSE, sep='\t', col.names = TRUE, row.names = FALSE)
```
