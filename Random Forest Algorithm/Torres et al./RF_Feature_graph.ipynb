```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv("RF_plot_time1.csv")
df1.rename(columns={'Unnamed: 0':'OTU'}, inplace=True)
df2 = pd.read_csv("RF_plot_time2.csv")
df2.rename(columns={'Unnamed: 0':'OTU'}, inplace=True)
df3 = pd.read_csv("RF_plot_time3.csv")
df3.rename(columns={'Unnamed: 0':'OTU'}, inplace=True)
df4 = pd.read_csv("RF_plot_time4.csv")
df4.rename(columns={'Unnamed: 0':'OTU'}, inplace=True)
df5 = pd.read_csv("RF_plot_time5.csv")
df5.rename(columns={'Unnamed: 0':'OTU'}, inplace=True)
```

```python
fig,axes = plt.subplots(nrows= 3, ncols=2, figsize=(20,15))
fig.subplots_adjust(hspace=.3,wspace=0.65)
df1.plot(kind = 'barh',x = "OTU",y="Gini-importance",legend = False,ax = axes[0,0]).invert_yaxis();axes[0,0].set_title('TIME1');axes[0,0].set_xlim([0, 0.15])
df2.plot(kind = 'barh',x = "OTU",y="Gini-importance",legend = False,ax = axes[0,1]).invert_yaxis();axes[0,1].set_title('TIME2');axes[0,1].set_xlim([0, 0.15])
df3.plot(kind = 'barh',x = "OTU",y="Gini-importance",legend = False,ax = axes[1,0]).invert_yaxis();axes[1,0].set_title('TIME3');axes[1,0].set_xlim([0, 0.15])
df4.plot(kind = 'barh',x = "OTU",y="Gini-importance",legend = False,ax = axes[1,1]).invert_yaxis();axes[1,1].set_title('TIME4');axes[1,1].set_xlim([0, 0.15])
df5.plot(kind = 'barh',x = "OTU",y="Gini-importance",legend = False,ax = axes[2,0]).invert_yaxis();axes[2,0].set_title('TIME5');axes[2,0].set_xlim([0, 0.15])
fig.delaxes(axes[2,1])
fig.text(0.45, 0.087, 'Importance', ha='center', size = 20)
fig.text(-.045, 0.5, 'OTUs', va='center', rotation='vertical', size = 20)   
fig.align_ylabels(axes[:, 0])
fig.align_ylabels(axes[:, 1])
plt.savefig("RF_Features_importance.pdf",format='pdf', dpi=650,bbox_inches = "tight")
```
